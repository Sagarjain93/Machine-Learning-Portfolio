# **Customer Churn in Telecom**

## **Problem Statement**

**Objective**

*  Predict whether a telecom customer will churn (Yes/No).

*  Identify key factors influencing churn.

*  Use insights to help business reduce customer loss.

**Business Understanding**

*  Churn = revenue loss.

*  Goal: Find high-risk customers and retention drivers.

**ML Type**

Binary Classification

Target: Churn
